```python
#Bai9
import os

# --- C·∫§U H√åNH T·ªÜP D·ªÆ LI·ªÜU ---
DATA_FILE_DM = 'danh_muc.txt' # T·ªáp l∆∞u danh m·ª•c
DATA_FILE_SP = 'san_pham.txt' # T·ªáp l∆∞u s·∫£n ph·∫©m
DELIMITER = '|' 

# --- C·∫§U TR√öC D·ªÆ LI·ªÜU ƒê∆†N GI·∫¢N (GLOBAL) ---
# Danh s√°ch ch·ª©a c√°c dictionary: [{'ma': 'TH', 'ten': 'Th·ªùi trang'}, ...]
DANH_MUC = [] 
# Danh s√°ch ch·ª©a c√°c dictionary: [{'ma': 'IP12', 'ten': 'iPhone 12', 'gia': 18000000.0, 'ma_dm': 'DT'}, ...]
SAN_PHAM = [] 

# ====================================================================
# --- C√ÅC H√ÄM X·ª¨ L√ù T·ªÜP V√Ä D·ªÆ LI·ªÜU ---
# ====================================================================

def doc_du_lieu():
    """ƒê·ªçc d·ªØ li·ªáu t·ª´ hai t·ªáp TEXT: danh_muc.txt v√† san_pham.txt."""
    global DANH_MUC, SAN_PHAM
    
    # 1. ƒê·ªçc Danh M·ª•c
    if os.path.exists(DATA_FILE_DM):
        try:
            with open(DATA_FILE_DM, 'r', encoding='utf-8') as f:
                lines = f.readlines()
            
            temp_dm = []
            for line in lines:
                line = line.strip()
                if not line: continue
                parts = line.split(DELIMITER)
                if len(parts) == 2:
                    temp_dm.append({'ma': parts[0], 'ten': parts[1]})
            DANH_MUC = temp_dm
            print(f"‚úÖ ƒê√£ ƒë·ªçc {len(DANH_MUC)} Danh m·ª•c.")
        except Exception as e:
            print(f"‚ùå L·ªói khi ƒë·ªçc t·ªáp Danh m·ª•c: {e}")
    else:
        print(f"üì£ T·ªáp {DATA_FILE_DM} kh√¥ng t·ªìn t·∫°i. DM tr·ªëng.")

    # 2. ƒê·ªçc S·∫£n Ph·∫©m
    if os.path.exists(DATA_FILE_SP):
        try:
            with open(DATA_FILE_SP, 'r', encoding='utf-8') as f:
                lines = f.readlines()
            
            temp_sp = []
            for line in lines:
                line = line.strip()
                if not line: continue
                parts = line.split(DELIMITER)
                if len(parts) == 4:
                    try:
                        temp_sp.append({
                            'ma': parts[0], 
                            'ten': parts[1], 
                            'gia': float(parts[2]), 
                            'ma_dm': parts[3]
                        })
                    except ValueError:
                        print(f"L·ªói: Gi√° kh√¥ng h·ª£p l·ªá cho s·∫£n ph·∫©m c√≥ m√£ {parts[0]}")
            SAN_PHAM = temp_sp
            print(f"‚úÖ ƒê√£ ƒë·ªçc {len(SAN_PHAM)} S·∫£n ph·∫©m.")
        except Exception as e:
            print(f"‚ùå L·ªói khi ƒë·ªçc t·ªáp S·∫£n ph·∫©m: {e}")
    else:
        print(f"üì£ T·ªáp {DATA_FILE_SP} kh√¥ng t·ªìn t·∫°i. SP tr·ªëng.")

def luu_du_lieu():
    """Ghi d·ªØ li·ªáu t·ª´ DANH_MUC v√† SAN_PHAM v√†o hai t·ªáp TEXT."""
    global DANH_MUC, SAN_PHAM
    
    # 1. Ghi Danh M·ª•c
    try:
        with open(DATA_FILE_DM, 'w', encoding='utf-8') as f:
            for dm in DANH_MUC:
                f.write(f"{dm['ma']}{DELIMITER}{dm['ten']}\n")
        print(f"‚úÖ ƒê√£ l∆∞u {len(DANH_MUC)} Danh m·ª•c v√†o {DATA_FILE_DM}.")
    except Exception as e:
        print(f"‚ùå L·ªói khi l∆∞u t·ªáp Danh m·ª•c: {e}")

    # 2. Ghi S·∫£n Ph·∫©m
    try:
        with open(DATA_FILE_SP, 'w', encoding='utf-8') as f:
            for sp in SAN_PHAM:
                f.write(f"{sp['ma']}{DELIMITER}{sp['ten']}{DELIMITER}{sp['gia']}{DELIMITER}{sp['ma_dm']}\n")
        print(f"‚úÖ ƒê√£ l∆∞u {len(SAN_PHAM)} S·∫£n ph·∫©m v√†o {DATA_FILE_SP}.")
    except Exception as e:
        print(f"‚ùå L·ªói khi l∆∞u t·ªáp S·∫£n ph·∫©m: {e}")

# --- C√ÅC H√ÄM H·ªñ TR·ª¢ CRUD ---

def tim_dm_theo_ma(ma):
    """T√¨m danh m·ª•c trong DANH_MUC theo m√£."""
    for dm in DANH_MUC:
        if dm['ma'] == ma:
            return dm
    return None

def tim_sp_theo_ma(ma):
    """T√¨m s·∫£n ph·∫©m trong SAN_PHAM theo m√£."""
    for sp in SAN_PHAM:
        if sp['ma'] == ma:
            return sp
    return None

def dem_sp_thuoc_dm(ma_dm):
    """ƒê·∫øm s·ªë l∆∞·ª£ng s·∫£n ph·∫©m thu·ªôc m·ªôt danh m·ª•c."""
    count = 0
    for sp in SAN_PHAM:
        if sp['ma_dm'] == ma_dm:
            count += 1
    return count

# --- C√ÅC H√ÄM CH√çNH (CRUD V√Ä T√åM KI·∫æM/S·∫ÆP X·∫æP) ---

def them_dm():
    """Th√™m Danh m·ª•c m·ªõi."""
    ma = input("Nh·∫≠p M√£ DM: ").upper()
    if tim_dm_theo_ma(ma):
        print(f"‚ùå L·ªói: M√£ danh m·ª•c '{ma}' ƒë√£ t·ªìn t·∫°i.")
        return
    
    ten = input("Nh·∫≠p T√™n DM: ")
    DANH_MUC.append({'ma': ma, 'ten': ten})
    print(f"‚úÖ ƒê√£ th√™m danh m·ª•c m·ªõi: M√£={ma}, T√™n={ten}")

def them_sp():
    """Th√™m S·∫£n ph·∫©m m·ªõi."""
    ma = input("Nh·∫≠p M√£ SP: ").upper()
    if tim_sp_theo_ma(ma):
        print(f"‚ùå L·ªói: M√£ s·∫£n ph·∫©m '{ma}' ƒë√£ t·ªìn t·∫°i.")
        return
    
    ten = input("Nh·∫≠p T√™n SP: ")
    
    try:
        gia = float(input("Nh·∫≠p ƒê∆°n gi√°: "))
    except ValueError:
        print("‚ùå L·ªói: ƒê∆°n gi√° ph·∫£i l√† m·ªôt s·ªë.")
        return

    ma_dm = input("Nh·∫≠p M√£ DM thu·ªôc v·ªÅ: ").upper()
    if not tim_dm_theo_ma(ma_dm):
        print(f"‚ùå L·ªói: M√£ danh m·ª•c '{ma_dm}' kh√¥ng t·ªìn t·∫°i. Vui l√≤ng th√™m DM tr∆∞·ªõc.")
        return
            
    SAN_PHAM.append({'ma': ma, 'ten': ten, 'gia': gia, 'ma_dm': ma_dm})
    print(f"‚úÖ ƒê√£ th√™m s·∫£n ph·∫©m m·ªõi: M√£={ma}, T√™n={ten}, Gi√°={gia:,.0f} VNƒê")

def sua_du_lieu():
    """S·ª≠a th√¥ng tin Danh m·ª•c ho·∫∑c S·∫£n ph·∫©m."""
    item_type = input("S·ª≠a (D)anh m·ª•c hay (S)·∫£n ph·∫©m? (D/S): ").upper()
    ma = input(f"Nh·∫≠p M√£ {item_type} c·∫ßn s·ª≠a: ").upper()
    
    if item_type == 'D':
        dm = tim_dm_theo_ma(ma)
        if not dm:
            print(f"‚ùå L·ªói: M√£ danh m·ª•c '{ma}' kh√¥ng t·ªìn t·∫°i.")
            return
        new_ten = input(f"Nh·∫≠p T√™n Danh m·ª•c m·ªõi (hi·ªán t·∫°i: {dm['ten']}): ")
        dm['ten'] = new_ten
        print(f"‚úÖ ƒê√£ c·∫≠p nh·∫≠t T√™n danh m·ª•c '{ma}' th√†nh '{new_ten}'.")
        
    elif item_type == 'S':
        sp = tim_sp_theo_ma(ma)
        if not sp:
            print(f"‚ùå L·ªói: M√£ s·∫£n ph·∫©m '{ma}' kh√¥ng t·ªìn t·∫°i.")
            return
        
        print(f"--- Th√¥ng tin SP hi·ªán t·∫°i: {sp['ten']} | {sp['gia']:,.0f} VNƒê | DM: {sp['ma_dm']}")
        
        new_ten = input("Nh·∫≠p T√™n SP m·ªõi (B·ªè qua n·∫øu kh√¥ng ƒë·ªïi): ")
        if new_ten:
            sp['ten'] = new_ten
            
        new_gia_str = input("Nh·∫≠p Gi√° m·ªõi (B·ªè qua n·∫øu kh√¥ng ƒë·ªïi): ")
        if new_gia_str:
            try:
                sp['gia'] = float(new_gia_str)
            except ValueError:
                print("‚ùå L·ªói: Gi√° m·ªõi kh√¥ng h·ª£p l·ªá. Thao t√°c s·ª≠a gi√° b·ªã h·ªßy.")
                return

        new_ma_dm = input("Nh·∫≠p M√£ DM m·ªõi (B·ªè qua n·∫øu kh√¥ng ƒë·ªïi): ").upper()
        if new_ma_dm and new_ma_dm != sp['ma_dm']:
            if not tim_dm_theo_ma(new_ma_dm):
                print(f"‚ùå L·ªói: M√£ danh m·ª•c m·ªõi '{new_ma_dm}' kh√¥ng t·ªìn t·∫°i.")
                return
            sp['ma_dm'] = new_ma_dm
            print(f"üîÑ ƒê√£ chuy·ªÉn SP '{ma}' sang DM '{new_ma_dm}'.")

        print(f"‚úÖ ƒê√£ c·∫≠p nh·∫≠t SP '{ma}'.")
    else:
        print("L·ª±a ch·ªçn kh√¥ng h·ª£p l·ªá.")

def xoa_du_lieu():
    """X√≥a Danh m·ª•c ho·∫∑c S·∫£n ph·∫©m."""
    global DANH_MUC, SAN_PHAM
    item_type = input("X√≥a (D)anh m·ª•c hay (S)·∫£n ph·∫©m? (D/S): ").upper()
    ma = input(f"Nh·∫≠p M√£ {item_type} c·∫ßn x√≥a: ").upper()
    
    if item_type == 'D':
        dm = tim_dm_theo_ma(ma)
        if not dm:
            print(f"‚ùå L·ªói: M√£ danh m·ª•c '{ma}' kh√¥ng t·ªìn t·∫°i.")
            return

        # L·ªçc l·∫°i danh s√°ch s·∫£n ph·∫©m, gi·ªØ l·∫°i nh·ªØng SP kh√¥ng thu·ªôc DM n√†y
        sp_truoc = len(SAN_PHAM)
        SAN_PHAM = [sp for sp in SAN_PHAM if sp['ma_dm'] != ma]
        sp_da_xoa = sp_truoc - len(SAN_PHAM)
        
        # L·ªçc l·∫°i danh s√°ch danh m·ª•c, x√≥a DM n√†y
        DANH_MUC = [dm for dm in DANH_MUC if dm['ma'] != ma]
        
        print(f"‚úÖ ƒê√£ x√≥a Danh m·ª•c '{ma}' v√† {sp_da_xoa} s·∫£n ph·∫©m li√™n quan.")

    elif item_type == 'S':
        sp = tim_sp_theo_ma(ma)
        if not sp:
            print(f"‚ùå L·ªói: M√£ s·∫£n ph·∫©m '{ma}' kh√¥ng t·ªìn t·∫°i.")
            return
        
        # L·ªçc l·∫°i danh s√°ch s·∫£n ph·∫©m, gi·ªØ l·∫°i nh·ªØng SP c√≥ m√£ kh√°c
        SAN_PHAM = [sp_item for sp_item in SAN_PHAM if sp_item['ma'] != ma]
        print(f"‚úÖ ƒê√£ x√≥a S·∫£n ph·∫©m '{ma}' - {sp['ten']}.")
        
    else:
        print("L·ª±a ch·ªçn kh√¥ng h·ª£p l·ªá.")

def tim_kiem_sp():
    """T√¨m ki·∫øm s·∫£n ph·∫©m theo T√™n ho·∫∑c M√£."""
    keyword = input("Nh·∫≠p t·ª´ kh√≥a (M√£ ho·∫∑c T√™n SP): ").lower().strip()
    results = []
    
    for sp in SAN_PHAM:
        if keyword in sp['ten'].lower() or keyword in sp['ma'].lower():
            results.append(sp)
            
    print(f"\nüîé T√¨m th·∫•y {len(results)} k·∫øt qu·∫£:")
    for sp in results:
        print(f"  [SP] M√£: {sp['ma']} | T√™n: {sp['ten']:<15} | Gi√°: {sp['gia']:,.0f} VNƒê | DM: {sp['ma_dm']}")
    
def hien_thi_all():
    """Hi·ªÉn th·ªã t·∫•t c·∫£ danh m·ª•c v√† s·∫£n ph·∫©m (s·∫Øp x·∫øp theo m√£ danh m·ª•c)."""
    
    # 1. S·∫Øp x·∫øp Danh m·ª•c theo m√£ (s·ª≠ d·ª•ng sorted() v·ªõi lambda)
    dm_sorted = sorted(DANH_MUC, key=lambda dm: dm['ma'])
    
    print("-" * 60)
    print(f"T·ªîNG QUAN D·ªÆ LI·ªÜU ({len(DANH_MUC)} DM, {len(SAN_PHAM)} SP)")
    print("-" * 60)
    
    for dm in dm_sorted:
        ma_dm = dm['ma']
        ten_dm = dm['ten']
        
        # ƒê·∫øm v√† in Danh m·ª•c
        so_luong_sp = dem_sp_thuoc_dm(ma_dm)
        print(f"[DM] M√£: {ma_dm} | T√™n: {ten_dm:<20} | S·ªë l∆∞·ª£ng SP: {so_luong_sp}")
        
        # L·ªçc v√† S·∫Øp x·∫øp S·∫£n ph·∫©m thu·ªôc Danh m·ª•c n√†y (s·∫Øp x·∫øp theo T√™n SP)
        sp_trong_dm = [sp for sp in SAN_PHAM if sp['ma_dm'] == ma_dm]
        sp_sorted = sorted(sp_trong_dm, key=lambda sp: sp['ten'].lower())
        
        if not sp_sorted:
            print("  (Danh m·ª•c n√†y ch∆∞a c√≥ s·∫£n ph·∫©m n√†o)")
        else:
            for sp in sp_sorted:
                print(f"  [SP] M√£: {sp['ma']} | T√™n: {sp['ten']:<15} | Gi√°: {sp['gia']:,.0f} VNƒê")
        print("-" * 60)

# ====================================================================
# --- CH∆Ø∆†NG TR√åNH CH√çNH (MAIN FUNCTION) ---
# ====================================================================

def main():
    doc_du_lieu() # T·ª± ƒë·ªông ƒë·ªçc d·ªØ li·ªáu khi kh·ªüi ƒë·ªông
    
    while True:
        print("\n" * 2)
        print("=" * 40)
        print("QU·∫¢N L√ù S·∫¢N PH·∫®M (CODE ƒê∆†N GI·∫¢N)")
        print("=" * 40)
        print("1. Th√™m m·ªõi Danh m·ª•c")
        print("2. Th√™m m·ªõi S·∫£n ph·∫©m (Y√™u c·∫ßu DM ƒë√£ t·ªìn t·∫°i)")
        print("3. S·ª≠a/X√≥a d·ªØ li·ªáu")
        print("4. Hi·ªÉn th·ªã t·∫•t c·∫£ SP theo DM (C√≥ s·∫Øp x·∫øp theo T√™n SP)")
        print("5. T√¨m ki·∫øm S·∫£n ph·∫©m (theo M√£ ho·∫∑c T√™n)")
        print("6. L∆∞u d·ªØ li·ªáu v√† Tho√°t")
        print("7. Ch·ªâ Tho√°t (Kh√¥ng L∆∞u)")
        print("-" * 40)
        
        choice = input("Nh·∫≠p l·ª±a ch·ªçn c·ªßa b·∫°n (1-7): ")

        if choice == '1':
            them_dm()
        elif choice == '2':
            them_sp()
        elif choice == '3':
            sua_xoa_choice = input("B·∫°n mu·ªën (1) S·ª≠a hay (2) X√≥a? (1/2): ")
            if sua_xoa_choice == '1':
                sua_du_lieu()
            elif sua_xoa_choice == '2':
                xoa_du_lieu()
            else:
                print("L·ª±a ch·ªçn kh√¥ng h·ª£p l·ªá.")
        elif choice == '4':
            hien_thi_all()
        elif choice == '5':
            tim_kiem_sp()
        elif choice == '6':
            luu_du_lieu()
            print("Tho√°t ch∆∞∆°ng tr√¨nh. T·∫°m bi·ªát!")
            break
        elif choice == '7':
            print("Tho√°t ch∆∞∆°ng tr√¨nh m√† kh√¥ng l∆∞u. T·∫°m bi·ªát!")
            break
        else:
            print("L·ª±a ch·ªçn kh√¥ng h·ª£p l·ªá. Vui l√≤ng nh·∫≠p s·ªë t·ª´ 1 ƒë·∫øn 7.")

if __name__ == "__main__":
    main()
```

```python
#Bai10
import json
import os

# --- C·∫§U H√åNH T·ªÜP D·ªÆ LI·ªÜU ---
DATA_FILE = 'students_data.json'

# --- C·∫§U TR√öC D·ªÆ LI·ªÜU CH√çNH (GLOBAL) ---
# ƒê√¢y l√† danh s√°ch c√°c l·ªõp h·ªçc, m·ªói l·ªõp l√† m·ªôt dictionary
CLASSES = [] 

# ====================================================================
# --- C√ÅC H√ÄM X·ª¨ L√ù T·ªÜP JSON ---
# ====================================================================

def doc_du_lieu_json():
    """ƒê·ªçc d·ªØ li·ªáu t·ª´ t·ªáp JSON."""
    global CLASSES
    
    if not os.path.exists(DATA_FILE):
        print("üì£ T·ªáp d·ªØ li·ªáu kh√¥ng t·ªìn t·∫°i. B·∫Øt ƒë·∫ßu v·ªõi d·ªØ li·ªáu tr·ªëng.")
        return

    try:
        with open(DATA_FILE, 'r', encoding='utf-8') as f:
            CLASSES = json.load(f)
        print(f"‚úÖ ƒê√£ ƒë·ªçc d·ªØ li·ªáu th√†nh c√¥ng t·ª´ {DATA_FILE}. T·ªïng {len(CLASSES)} l·ªõp h·ªçc.")
    except json.JSONDecodeError:
        print("‚ùå L·ªói: T·ªáp JSON kh√¥ng h·ª£p l·ªá. Kh·ªüi t·∫°o d·ªØ li·ªáu tr·ªëng.")
        CLASSES = []
    except Exception as e:
        print(f"‚ùå L·ªói khi ƒë·ªçc t·ªáp d·ªØ li·ªáu: {e}")

def luu_du_lieu_json():
    """Ghi d·ªØ li·ªáu v√†o t·ªáp JSON."""
    global CLASSES
    try:
        with open(DATA_FILE, 'w', encoding='utf-8') as f:
            # Ghi d·ªØ li·ªáu v·ªõi indent=4 ƒë·ªÉ d·ªÖ ƒë·ªçc h∆°n
            json.dump(CLASSES, f, ensure_ascii=False, indent=4) 
        print(f"‚úÖ ƒê√£ l∆∞u t·∫•t c·∫£ d·ªØ li·ªáu th√†nh c√¥ng v√†o {DATA_FILE}.")
    except Exception as e:
        print(f"‚ùå L·ªói khi l∆∞u t·ªáp d·ªØ li·ªáu: {e}")

# ====================================================================
# --- C√ÅC H√ÄM H·ªñ TR·ª¢ T√åM KI·∫æM ---
# ====================================================================

def tim_lop_theo_ma(ma_lop):
    """T√¨m dictionary L·ªõp h·ªçc theo m√£."""
    for lop in CLASSES:
        if lop['ma_lop'].upper() == ma_lop.upper():
            return lop
    return None

def tim_sv_theo_ma(ma_sv):
    """T√¨m dictionary Sinh vi√™n theo m√£ trong t·∫•t c·∫£ c√°c l·ªõp."""
    for lop in CLASSES:
        for sv in lop['sinh_vien']:
            if sv['ma_sv'].upper() == ma_sv.upper():
                # Tr·∫£ v·ªÅ c·∫£ sinh vi√™n v√† l·ªõp ch·ª©a sinh vi√™n ƒë√≥
                return sv, lop
    return None, None

# ====================================================================
# --- C√ÅC H√ÄM CH√çNH (CRUD) ---
# ====================================================================

# --- 1. TH√äM M·ªöI ---
def them_lop():
    """Th√™m L·ªõp h·ªçc m·ªõi."""
    ma = input("Nh·∫≠p M√£ L·ªõp: ").upper()
    if tim_lop_theo_ma(ma):
        print(f"‚ùå L·ªói: M√£ l·ªõp '{ma}' ƒë√£ t·ªìn t·∫°i.")
        return
    
    ten = input("Nh·∫≠p T√™n L·ªõp: ")
    CLASSES.append({'ma_lop': ma, 'ten_lop': ten, 'sinh_vien': []})
    print(f"‚úÖ ƒê√£ th√™m l·ªõp m·ªõi: M√£={ma}, T√™n={ten}")

def them_sv():
    """Th√™m Sinh vi√™n m·ªõi."""
    ma_sv = input("Nh·∫≠p M√£ SV: ").upper()
    if tim_sv_theo_ma(ma_sv)[0]:
        print(f"‚ùå L·ªói: M√£ sinh vi√™n '{ma_sv}' ƒë√£ t·ªìn t·∫°i.")
        return
    
    ten_sv = input("Nh·∫≠p T√™n SV: ")
    
    try:
        nam_sinh = int(input("Nh·∫≠p NƒÉm sinh (VD: 2003): "))
    except ValueError:
        print("‚ùå L·ªói: NƒÉm sinh ph·∫£i l√† s·ªë nguy√™n.")
        return

    ma_lop = input("Nh·∫≠p M√£ L·ªõp thu·ªôc v·ªÅ: ").upper()
    lop = tim_lop_theo_ma(ma_lop)
    
    if not lop:
        print(f"‚ùå L·ªói: M√£ L·ªõp '{ma_lop}' kh√¥ng t·ªìn t·∫°i. Vui l√≤ng th√™m L·ªõp tr∆∞·ªõc.")
        return
            
    sv_moi = {'ma_sv': ma_sv, 'ten_sv': ten_sv, 'nam_sinh': nam_sinh}
    lop['sinh_vien'].append(sv_moi)
    print(f"‚úÖ ƒê√£ th√™m sinh vi√™n m·ªõi: {ten_sv} ({ma_sv}) v√†o l·ªõp {lop['ten_lop']}.")

# --- 2. S·ª¨A ---
def sua_du_lieu():
    """S·ª≠a th√¥ng tin L·ªõp ho·∫∑c Sinh vi√™n."""
    item_type = input("S·ª≠a (L)·ªõp hay (S)inh vi√™n? (L/S): ").upper()
    ma = input(f"Nh·∫≠p M√£ {item_type} c·∫ßn s·ª≠a: ").upper()
    
    if item_type == 'L':
        lop = tim_lop_theo_ma(ma)
        if not lop:
            print(f"‚ùå L·ªói: M√£ l·ªõp '{ma}' kh√¥ng t·ªìn t·∫°i.")
            return
        new_ten = input(f"Nh·∫≠p T√™n L·ªõp m·ªõi (hi·ªán t·∫°i: {lop['ten_lop']}): ")
        lop['ten_lop'] = new_ten
        print(f"‚úÖ ƒê√£ c·∫≠p nh·∫≠t T√™n l·ªõp '{ma}' th√†nh '{new_ten}'.")
        
    elif item_type == 'S':
        sv, lop = tim_sv_theo_ma(ma)
        if not sv:
            print(f"‚ùå L·ªói: M√£ sinh vi√™n '{ma}' kh√¥ng t·ªìn t·∫°i.")
            return
        
        print(f"--- Th√¥ng tin SV hi·ªán t·∫°i: {sv['ten_sv']} | NƒÉm sinh: {sv['nam_sinh']} | L·ªõp: {lop['ten_lop']}")
        
        new_ten = input("Nh·∫≠p T√™n SV m·ªõi (B·ªè qua n·∫øu kh√¥ng ƒë·ªïi): ")
        if new_ten:
            sv['ten_sv'] = new_ten
            
        new_nam_sinh_str = input("Nh·∫≠p NƒÉm sinh m·ªõi (B·ªè qua n·∫øu kh√¥ng ƒë·ªïi): ")
        if new_nam_sinh_str:
            try:
                sv['nam_sinh'] = int(new_nam_sinh_str)
            except ValueError:
                print("‚ùå L·ªói: NƒÉm sinh m·ªõi kh√¥ng h·ª£p l·ªá. Thao t√°c s·ª≠a nƒÉm sinh b·ªã h·ªßy.")
                return

        print(f"‚úÖ ƒê√£ c·∫≠p nh·∫≠t SV '{ma}'.")
    else:
        print("L·ª±a ch·ªçn kh√¥ng h·ª£p l·ªá.")

# --- 3. X√ìA ---
def xoa_du_lieu():
    """X√≥a L·ªõp ho·∫∑c Sinh vi√™n."""
    global CLASSES
    item_type = input("X√≥a (L)·ªõp hay (S)inh vi√™n? (L/S): ").upper()
    ma = input(f"Nh·∫≠p M√£ {item_type} c·∫ßn x√≥a: ").upper()
    
    if item_type == 'L':
        lop = tim_lop_theo_ma(ma)
        if not lop:
            print(f"‚ùå L·ªói: M√£ l·ªõp '{ma}' kh√¥ng t·ªìn t·∫°i.")
            return
        
        sv_count = len(lop['sinh_vien'])
        
        # L·ªçc l·∫°i danh s√°ch l·ªõp, gi·ªØ l·∫°i nh·ªØng l·ªõp c√≥ m√£ kh√°c
        CLASSES[:] = [l for l in CLASSES if l['ma_lop'] != ma]
        
        print(f"‚úÖ ƒê√£ x√≥a L·ªõp '{ma}' v√† {sv_count} sinh vi√™n thu·ªôc l·ªõp ƒë√≥.")

    elif item_type == 'S':
        sv, lop = tim_sv_theo_ma(ma)
        if not sv:
            print(f"‚ùå L·ªói: M√£ sinh vi√™n '{ma}' kh√¥ng t·ªìn t·∫°i.")
            return
        
        # L·ªçc l·∫°i danh s√°ch sinh vi√™n trong l·ªõp ƒë√≥, x√≥a SV c√≥ m√£ c·∫ßn x√≥a
        lop['sinh_vien'][:] = [s for s in lop['sinh_vien'] if s['ma_sv'] != ma]
        print(f"‚úÖ ƒê√£ x√≥a Sinh vi√™n '{ma}' - {sv['ten_sv']} kh·ªèi l·ªõp {lop['ten_lop']}.")
        
    else:
        print("L·ª±a ch·ªçn kh√¥ng h·ª£p l·ªá.")

# --- 4. T√åM KI·∫æM V√Ä S·∫ÆP X·∫æP ---
def tim_kiem_sv():
    """T√¨m ki·∫øm sinh vi√™n theo T√™n ho·∫∑c M√£ trong t·∫•t c·∫£ c√°c l·ªõp."""
    keyword = input("Nh·∫≠p t·ª´ kh√≥a (M√£, T√™n SV ho·∫∑c NƒÉm sinh): ").lower().strip()
    results = []
    
    for lop in CLASSES:
        for sv in lop['sinh_vien']:
            search_str = f"{sv['ma_sv']} {sv['ten_sv']} {sv['nam_sinh']}".lower()
            if keyword in search_str:
                results.append((sv, lop)) # L∆∞u c·∫£ SV v√† L·ªõp
            
    print(f"\nüîé T√¨m th·∫•y {len(results)} k·∫øt qu·∫£:")
    for sv, lop in results:
        print(f"  [SV] M√£: {sv['ma_sv']} | T√™n: {sv['ten_sv']:<20} | NS: {sv['nam_sinh']} | L·ªõp: {lop['ten_lop']} ({lop['ma_lop']})")

def hien_thi_all():
    """Hi·ªÉn th·ªã t·∫•t c·∫£ l·ªõp v√† sinh vi√™n, c√≥ s·∫Øp x·∫øp."""
    
    # 1. S·∫Øp x·∫øp Danh s√°ch L·ªõp theo M√£
    classes_sorted = sorted(CLASSES, key=lambda l: l['ma_lop'])
    
    print("-" * 70)
    print(f"T·ªîNG QUAN D·ªÆ LI·ªÜU ({len(CLASSES)} L·ªöP)")
    print("-" * 70)
    
    for lop in classes_sorted:
        ma_lop = lop['ma_lop']
        ten_lop = lop['ten_lop']
        
        # 2. S·∫Øp x·∫øp Sinh vi√™n trong L·ªõp theo T√™n
        # T√™n SV l√† thu·ªôc t√≠nh ƒë∆∞·ª£c ∆∞u ti√™n s·∫Øp x·∫øp
        sv_sorted = sorted(lop['sinh_vien'], key=lambda sv: sv['ten_sv'].lower())
        
        print(f"[L·ªöP] M√£: {ma_lop} | T√™n: {ten_lop:<30} | S·ªë l∆∞·ª£ng SV: {len(sv_sorted)}")
        
        if not sv_sorted:
            print("  (L·ªõp n√†y ch∆∞a c√≥ sinh vi√™n n√†o)")
        else:
            for sv in sv_sorted:
                print(f"  [SV] M√£: {sv['ma_sv']} | T√™n: {sv['ten_sv']:<20} | NƒÉm sinh: {sv['nam_sinh']}")
        print("-" * 70)

# ====================================================================
# --- CH∆Ø∆†NG TR√åNH CH√çNH (MAIN FUNCTION) ---
# ====================================================================

def main():
    doc_du_lieu_json() # T·ª± ƒë·ªông ƒë·ªçc d·ªØ li·ªáu JSON khi kh·ªüi ƒë·ªông
    
    while True:
        print("\n" * 2)
        print("=" * 60)
        print("PH·∫¶N M·ªÄM QU·∫¢N L√ù SINH VI√äN (S·ª¨ D·ª§NG JSON)")
        print("=" * 60)
        print("1. Th√™m m·ªõi L·ªõp h·ªçc")
        print("2. Th√™m m·ªõi Sinh vi√™n")
        print("3. S·ª≠a th√¥ng tin (L·ªõp ho·∫∑c Sinh vi√™n)")
        print("4. X√≥a th√¥ng tin (L·ªõp ho·∫∑c Sinh vi√™n)")
        print("5. Hi·ªÉn th·ªã t·∫•t c·∫£ (L·ªõp theo M√£, SV theo T√™n)")
        print("6. T√¨m ki·∫øm Sinh vi√™n")
        print("7. L∆∞u d·ªØ li·ªáu v√† Tho√°t")
        print("8. Ch·ªâ Tho√°t (Kh√¥ng L∆∞u)")
        print("-" * 60)
        
        choice = input("Nh·∫≠p l·ª±a ch·ªçn c·ªßa b·∫°n (1-8): ")

        if choice == '1':
            them_lop()
        elif choice == '2':
            them_sv()
        elif choice == '3':
            sua_du_lieu()
        elif choice == '4':
            xoa_du_lieu()
        elif choice == '5':
            hien_thi_all()
        elif choice == '6':
            tim_kiem_sv()
        elif choice == '7':
            luu_du_lieu_json()
            print("Tho√°t ch∆∞∆°ng tr√¨nh. T·∫°m bi·ªát!")
            break
        elif choice == '8':
            print("Tho√°t ch∆∞∆°ng tr√¨nh m√† kh√¥ng l∆∞u. T·∫°m bi·ªát!")
            break
        else:
            print("L·ª±a ch·ªçn kh√¥ng h·ª£p l·ªá. Vui l√≤ng nh·∫≠p s·ªë t·ª´ 1 ƒë·∫øn 8.")

if __name__ == "__main__":
    main()
```

```python
#Bai11
import pandas as pd
import os # D√πng ƒë·ªÉ ki·ªÉm tra file t·ªìn t·∫°i

TEN_FILE = 'NhanVien.xlsx'

# --- 1. T·∫°o ho·∫∑c ƒê·ªçc File ---
def doc_du_lieu(ten_file):
    """ƒê·ªçc d·ªØ li·ªáu t·ª´ file Excel, n·∫øu file ch∆∞a c√≥ th√¨ tr·∫£ v·ªÅ DataFrame r·ªóng."""
    if os.path.exists(ten_file):
        # ƒê·∫£m b·∫£o c·ªôt Tu·ªïi l√† s·ªë nguy√™n khi ƒë·ªçc
        df = pd.read_excel(ten_file, dtype={'Tu·ªïi': int})
        return df
    else:
        # T·∫°o DataFrame r·ªóng v·ªõi c·∫•u tr√∫c mong mu·ªën
        return pd.DataFrame(columns=['STT', 'M√£', 'T√™n', 'Tu·ªïi'])

# --- 2. L∆∞u Nh√¢n Vi√™n (Th√™m m·ªõi) ---
def luu_nhan_vien(ma, ten, tuoi):
    df = doc_du_lieu(TEN_FILE)
    
    # T·∫°o d·ªØ li·ªáu m·ªõi
    du_lieu_moi = {'M√£': ma, 'T√™n': ten, 'Tu·ªïi': int(tuoi)}
    
    # Th√™m d·ªØ li·ªáu m·ªõi v√†o DataFrame
    # ignore_index=True ƒë·ªÉ t·ª± ƒë·ªông th√™m index m·ªõi
    df = pd.concat([df, pd.DataFrame([du_lieu_moi])], ignore_index=True)
    
    # C·∫≠p nh·∫≠t c·ªôt STT (STT = index + 1)
    df['STT'] = df.index + 1
    
    # L∆∞u l·∫°i v√†o file Excel
    df.to_excel(TEN_FILE, index=False) # index=False ƒë·ªÉ kh√¥ng ghi c·ªôt index c·ªßa pandas v√†o file
    print(f"‚úÖ ƒê√£ th√™m nh√¢n vi√™n {ten} ({ma}) v√†o file.")

# --- 3. ƒê·ªçc v√† Hi·ªÉn Th·ªã Danh S√°ch ---
def doc_danh_sach():
    df = doc_du_lieu(TEN_FILE)
    if df.empty:
        print("‚ùå Danh s√°ch nh√¢n vi√™n tr·ªëng.")
    else:
        print("\n=== üìã DANH S√ÅCH NH√ÇN VI√äN HI·ªÜN T·∫†I ===")
        print(df.to_string(index=False)) # Hi·ªÉn th·ªã kh√¥ng k√®m index c·ªßa pandas
        print("========================================\n")

# --- 4. S·∫Øp x·∫øp Nh√¢n vi√™n theo Tu·ªïi tƒÉng d·∫ßn ---
def sap_xep_theo_tuoi():
    df = doc_du_lieu(TEN_FILE)
    if df.empty:
        print("‚ùå Kh√¥ng c√≥ d·ªØ li·ªáu ƒë·ªÉ s·∫Øp x·∫øp.")
        return

    # S·∫Øp x·∫øp theo c·ªôt 'Tu·ªïi' tƒÉng d·∫ßn (ascending=True)
    df_sorted = df.sort_values(by='Tu·ªïi', ascending=True)
    
    # C·∫≠p nh·∫≠t l·∫°i c·ªôt STT theo th·ª© t·ª± m·ªõi sau khi s·∫Øp x·∫øp
    df_sorted['STT'] = range(1, len(df_sorted) + 1)
    
    print("\n=== üìà DANH S√ÅCH NH√ÇN VI√äN S·∫ÆP X·∫æP THEO TU·ªîI TƒÇNG D·∫¶N ===")
    print(df_sorted.to_string(index=False))
    print("==============================================================\n")
    
    # T√πy ch·ªçn: Ghi k·∫øt qu·∫£ s·∫Øp x·∫øp l·∫°i v√†o file (n·∫øu c·∫ßn)
    # df_sorted.to_excel(TEN_FILE, index=False)
    # print("‚úÖ ƒê√£ c·∫≠p nh·∫≠t file Excel theo th·ª© t·ª± tu·ªïi tƒÉng d·∫ßn.")


# --- V√ç D·ª§ S·ª¨ D·ª§NG C√ÅC H√ÄM ---
# 1. Th√™m m·ªôt v√†i nh√¢n vi√™n
luu_nhan_vien('NV7', 'Kh√°nh', 28)
luu_nhan_vien('NV8', 'Minh', 21)

# 2. ƒê·ªçc danh s√°ch
doc_danh_sach()

# 3. S·∫Øp x·∫øp v√† in ra
sap_xep_theo_tuoi()
```

```python
#Bai13
NHOM_THIET_BI = {
    'n1': 'Nh√≥m 1',
    'n2': 'Nh√≥m 2',
    'n3': 'Nh√≥m 3',
    # ...
}
DANH_SACH_THIET_BI = [
    {'ma': 'tb1', 'ten': 'Thi·∫øt b·ªã 1', 'manhom': 'n1'},
    {'ma': 'tb2', 'ten': 'Thi·∫øt b·ªã 2', 'manhom': 'n1'},
    {'ma': 'tb3', 'ten': 'Thi·∫øt b·ªã 3', 'manhom': 'n2'},
    # ...
]
import xml.etree.ElementTree as ET
from collections import defaultdict

# --- D·ªØ li·ªáu gi·∫£ l·∫≠p (thay cho vi·ªác ƒë·ªçc file) ---

# Gi·∫£ l·∫≠p d·ªØ li·ªáu t·ª´ nhomthietbi.xml
NHOM_THIET_BI_DICT = {
    'n1': 'Nh√≥m M√°y t√≠nh',
    'n2': 'Nh√≥m Thi·∫øt b·ªã m·∫°ng',
    'n3': 'Nh√≥m Thi·∫øt b·ªã ngo·∫°i vi',
    'n4': 'Nh√≥m Ph·∫ßn m·ªÅm',
}

# Gi·∫£ l·∫≠p d·ªØ li·ªáu t·ª´ ThietBi.xml
DANH_SACH_THIET_BI = [
    {'ma': 'tb1', 'ten': 'Laptop Dell', 'manhom': 'n1'},
    {'ma': 'tb2', 'ten': 'M√°y ch·ªß HP', 'manhom': 'n1'},
    {'ma': 'tb3', 'ten': 'Switch Cisco', 'manhom': 'n2'},
    {'ma': 'tb4', 'ten': 'Router TP-Link', 'manhom': 'n2'},
    {'ma': 'tb5', 'ten': 'Chu·ªôt Quang', 'manhom': 'n3'},
    {'ma': 'tb6', 'ten': 'Ph·∫ßn m·ªÅm K·∫ø to√°n', 'manhom': 'n4'},
    {'ma': 'tb7', 'ten': 'M√†n h√¨nh AOC', 'manhom': 'n3'},
    {'ma': 'tb8', 'ten': 'M√°y t√≠nh ƒë·ªÉ b√†n', 'manhom': 'n1'},
    {'ma': 'tb9', 'ten': 'Chu·ªôt kh√¥ng d√¢y', 'manhom': 'n3'},
]

# --- H√ÄM H·ªñ TR·ª¢ (N·∫øu b·∫°n c·∫ßn ƒë·ªçc XML th·∫≠t) ---
def doc_xml_thiet_bi(file_path):
    """
    H√†m m·∫´u ƒë·ªÉ ƒë·ªçc d·ªØ li·ªáu t·ª´ ThietBi.xml n·∫øu b·∫°n c√≥ file th·∫≠t.
    C·∫ßn ƒë·∫£m b·∫£o file XML c·ªßa b·∫°n tu√¢n th·ªß c·∫•u tr√∫c.
    """
    tree = ET.parse(file_path)
    root = tree.getroot()
    
    ds_thiet_bi = []
    # L·∫∑p qua t·∫•t c·∫£ c√°c th·∫ª <thietbi>
    for thietbi_element in root.findall('thietbi'):
        # L·∫•y manhom t·ª´ thu·ªôc t√≠nh (v√≠ d·ª• trong h√¨nh, manhom n·∫±m trong th·∫ª <thietbi manhom="n1">)
        # N·∫øu manhom l√† th·∫ª con: manhom_val = thietbi_element.find('manhom').text
        
        # D·ª±a tr√™n c·∫•u tr√∫c m·∫´u trong h√¨nh: <thietbi manhom="n1">
        manhom_val = thietbi_element.attrib.get('manhom')
        
        ds_thiet_bi.append({
            'manhom': manhom_val,
            'ma': thietbi_element.find('ma').text,
            'ten': thietbi_element.find('ten').text
        })
    return ds_thiet_bi

# --- 1. HI·ªÇN TH·ªä DANH S√ÅCH NH√ìM THI·∫æT B·ªä ---
def hien_thi_nhom_thiet_bi():
    """Hi·ªÉn th·ªã M√£ v√† T√™n c·ªßa t·∫•t c·∫£ c√°c nh√≥m thi·∫øt b·ªã."""
    print("\n=== 1Ô∏è‚É£ DANH S√ÅCH NH√ìM THI·∫æT B·ªä ===")
    print(f"{'M√£ Nh√≥m':<10} | {'T√™n Nh√≥m'}")
    print("-" * 30)
    for ma, ten in NHOM_THIET_BI_DICT.items():
        print(f"{ma:<10} | {ten}")
    print("====================================")

# --- 2. HI·ªÇN TH·ªä TO√ÄN B·ªò THI·∫æT B·ªä ---
def hien_thi_toan_bo_thiet_bi():
    """Hi·ªÉn th·ªã t·∫•t c·∫£ thi·∫øt b·ªã c√πng v·ªõi T√™n Nh√≥m t∆∞∆°ng ·ª©ng."""
    print("\n=== 2Ô∏è‚É£ TO√ÄN B·ªò DANH S√ÅCH THI·∫æT B·ªä ===")
    print(f"{'M√£ TB':<8} | {'T√™n Thi·∫øt B·ªã':<20} | {'M√£ Nh√≥m':<8} | {'T√™n Nh√≥m'}")
    print("-" * 60)
    for tb in DANH_SACH_THIET_BI:
        # L·∫•y T√™n Nh√≥m t·ª´ Dictionary
        ten_nhom = NHOM_THIET_BI_DICT.get(tb['manhom'], "Kh√¥ng x√°c ƒë·ªãnh")
        print(f"{tb['ma']:<8} | {tb['ten']:<20} | {tb['manhom']:<8} | {ten_nhom}")
    print("===============================================================")

# --- 3. L·ªåC DANH S√ÅCH THI·∫æT B·ªä THEO NH√ìM THI·∫æT B·ªä ---
def loc_thiet_bi_theo_nhom(ma_nhom_can_loc):
    """L·ªçc v√† hi·ªÉn th·ªã c√°c thi·∫øt b·ªã thu·ªôc m·ªôt m√£ nh√≥m c·ª• th·ªÉ."""
    ma_nhom_can_loc = ma_nhom_can_loc.lower()
    ten_nhom = NHOM_THIET_BI_DICT.get(ma_nhom_can_loc, "Kh√¥ng x√°c ƒë·ªãnh")
    
    print(f"\n=== 3Ô∏è‚É£ DANH S√ÅCH THI·∫æT B·ªä C·ª¶A NH√ìM '{ten_nhom}' ({ma_nhom_can_loc}) ===")
    
    # L·ªçc danh s√°ch
    ds_loc = [tb for tb in DANH_SACH_THIET_BI if tb['manhom'] == ma_nhom_can_loc]
    
    if not ds_loc:
        print(f"‚ùå Kh√¥ng t√¨m th·∫•y thi·∫øt b·ªã n√†o thu·ªôc nh√≥m '{ten_nhom}'.")
        return
        
    print(f"{'M√£ TB':<8} | {'T√™n Thi·∫øt B·ªã'}")
    print("-" * 30)
    for tb in ds_loc:
        print(f"{tb['ma']:<8} | {tb['ten']}")
    print("====================================")


# --- 4. XU·∫§T NH√ìM THI·∫æT B·ªä C√ì S·ªê L∆Ø·ª¢NG THI·∫æT B·ªä NHI·ªÄU NH·∫§T ---
def xuat_nhom_co_so_luong_nhieu_nhat():
    """ƒê·∫øm s·ªë l∆∞·ª£ng thi·∫øt b·ªã trong m·ªói nh√≥m v√† t√¨m nh√≥m c√≥ s·ªë l∆∞·ª£ng l·ªõn nh·∫•t."""
    
    # 1. ƒê·∫øm s·ªë l∆∞·ª£ng thi·∫øt b·ªã cho m·ªói nh√≥m
    dem_so_luong = defaultdict(int)
    for tb in DANH_SACH_THIET_BI:
        dem_so_luong[tb['manhom']] += 1
        
    if not dem_so_luong:
        print("\n‚ùå Danh s√°ch thi·∫øt b·ªã r·ªóng, kh√¥ng th·ªÉ x√°c ƒë·ªãnh nh√≥m c√≥ s·ªë l∆∞·ª£ng nhi·ªÅu nh·∫•t.")
        return
        
    # 2. T√¨m m√£ nh√≥m c√≥ s·ªë l∆∞·ª£ng l·ªõn nh·∫•t
    # max() v·ªõi key=dem_so_luong.get s·∫Ω t√¨m key (m√£ nh√≥m) c√≥ gi√° tr·ªã (s·ªë l∆∞·ª£ng) l·ªõn nh·∫•t
    ma_nhom_max = max(dem_so_luong, key=dem_so_luong.get)
    so_luong_max = dem_so_luong[ma_nhom_max]
    ten_nhom_max = NHOM_THIET_BI_DICT.get(ma_nhom_max, "Kh√¥ng x√°c ƒë·ªãnh")
    
    # 3. T√¨m c√°c nh√≥m c√≥ c√πng s·ªë l∆∞·ª£ng t·ªëi ƒëa (n·∫øu c√≥)
    nhom_max = [
        (ma, NHOM_THIET_BI_DICT.get(ma, "Kh√¥ng x√°c ƒë·ªãnh"), so_luong_max)
        for ma, so_luong in dem_so_luong.items() if so_luong == so_luong_max
    ]

    print("\n=== 4Ô∏è‚É£ NH√ìM C√ì S·ªê L∆Ø·ª¢NG THI·∫æT B·ªä NHI·ªÄU NH·∫§T ===")
    for ma, ten, sl in nhom_max:
         print(f"ü•á Nh√≥m: **{ten}** (M√£: {ma}) v·ªõi **{sl}** thi·∫øt b·ªã.")
    print("=========================================================")

# --- CH·∫†Y CH∆Ø∆†NG TR√åNH ---
# 1. Hi·ªÉn th·ªã danh s√°ch nh√≥m
hien_thi_nhom_thiet_bi()

# 2. Hi·ªÉn th·ªã to√†n b·ªô thi·∫øt b·ªã
hien_thi_toan_bo_thiet_bi()

# 3. L·ªçc danh s√°ch thi·∫øt b·ªã theo m·ªôt nh√≥m c·ª• th·ªÉ (V√≠ d·ª•: n1)
loc_thiet_bi_theo_nhom('n1')

# 4. Xu·∫•t nh√≥m c√≥ s·ªë l∆∞·ª£ng thi·∫øt b·ªã nhi·ªÅu nh·∫•t
xuat_nhom_co_so_luong_nhieu_nhat()
```
