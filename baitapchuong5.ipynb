```python
#Bai5
def xu_ly_chuoi():
    input_string = input("Moi nhap mot chuoi bat ky: ")
    
    dem_hoa = 0
    dem_thuong = 0
    dem_so = 0
    dem_khoang_trang = 0
    dem_nguyen_am = 0
    dem_phu_am = 0
    dem_ky_tu_dac_biet = 0
    
    nguyen_am_set = {'a', 'e', 'i', 'o', 'u', 'A', 'E', 'I', 'O', 'U'}
    
    for char in input_string:
        
        if char.isalpha():
            if char.isupper():
                dem_hoa += 1
            elif char.islower():
                dem_thuong += 1
            
            if char in nguyen_am_set:
                dem_nguyen_am += 1
            else:
                dem_phu_am += 1
            
        elif char.isdigit():
            dem_so += 1
            
        elif char.isspace():
            dem_khoang_trang += 1
            
        else:
            dem_ky_tu_dac_biet += 1

    print("\n--- KET QUA PHAN TICH CHUOI ---")
    print(f"Tong so ky tu trong chuoi: {len(input_string)}")
    print("---------------------------------")
    print(f"Bao nhieu chu IN HOA: {dem_hoa}")
    print(f"Bao nhieu chu in thuong: {dem_thuong}")
    print(f"Bao nhieu chu la chu so: {dem_so}")
    print(f"Bao nhieu chu la ky tu dac biet: {dem_ky_tu_dac_biet}")
    print(f"Bao nhieu chu la khoang trang: {dem_khoang_trang}")
    print(f"Bao nhieu chu la Nguyen Am: {dem_nguyen_am}")
    print(f"Bao nhieu chu la Phu Am: {dem_phu_am}")
    print("---------------------------------")

xu_ly_chuoi()
```

```python
#Bai6
import re

def NegativeNumberInStrings(str_data):
    negative_numbers_str = re.findall(r'-\d+', str_data)
    negative_numbers_int = [int(num) for num in negative_numbers_str]
    return negative_numbers_int

chuoi_mau = "abc-5xyz-12k9l--p"
ket_qua = NegativeNumberInStrings(chuoi_mau)

print(f"Chuỗi đầu vào: '{chuoi_mau}'")
print(f"Các số nguyên âm được trích lọc: {ket_qua}")
```

```python
#Bai7
def toi_uu_chuoi_danh_tu(chuoi_dau_vao):
    danh_sach_tu = chuoi_dau_vao.split()
    chuoi_da_xu_ly_khoang_trang = " ".join(danh_sach_tu)
    chuoi_toi_uu = chuoi_da_xu_ly_khoang_trang.title()
    
    return chuoi_toi_uu

chuoi_input = " TRần duY   thAnH "
chuoi_output = toi_uu_chuoi_danh_tu(chuoi_input)

print(f"Input: \"{chuoi_input}\"")
print(f"Output: \"{chuoi_output}\"")
```

```python
#Bai8
import os

def lay_ten_file_hoan_chinh(duong_dan):
    return os.path.basename(duong_dan)

def lay_ten_bai_hat(duong_dan):
    ten_file_hoan_chinh = os.path.basename(duong_dan)
    ten_bai_hat, _ = os.path.splitext(ten_file_hoan_chinh)
    return ten_bai_hat

duong_dan_test = r"d:\\music\muabui.mp3"

print(f"Đường dẫn đầu vào: {duong_dan_test}")
print(f"1. Tên file hoàn chỉnh: {lay_ten_file_hoan_chinh(duong_dan_test)}")
print(f"2. Tên bài hát: {lay_ten_bai_hat(duong_dan_test)}")
```
