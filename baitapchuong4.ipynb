```python
#Bai1
from math import sqrt
print("Chương trình tính diện tích Tam Giác")
a=float(input("Nhập cạnh a>0:"))
b=float(input("Nhập cạnh b>0:"))
c=float(input("Nhập cạnh c>0:"))
if (a<=0 or b <=0 or c <=0) or (a+b)<=c or (a+c)<=b or b+c<=a:
 print("Tam giác không hợp lệ")
else:
 cv=a+b+c
 p=cv/2
 dt=sqrt(p*(p-a)*(p-b)*(p-c))
 print("Diện tích =",dt)
```

```python
#Bai2
from random import randrange
while True:
    somay=randrange(1,101)
    solandoan=0
    win=False
    while solandoan<7:
        solandoan+=1
        songuoi=int(input("Máy đoán [1..100], mời bạn đoán:"))
        print("Bạn đoán lần thứ ",solandoan)
        if somay==songuoi:
            print("Chúc mừng bạn đoán đúng, số máy là=",somay)
            win=True
            break
        if somay>songuoi:
            print("Bạn đoán sai, số máy > số bạn")
        elif somay<songuoi:
            print("Bạn đoán sai, số máy < số bạn")
    if win==False:
            print("GAME OVER!, số máy =",somay)
    hoi=input("Tiếp không?")
    if hoi=="k":
        break
print("Cám ơn bạn đã chơi Game!")
```

```python
#Bai3
def BMI(height,weight):
 return weight/(height**2)
def PhanLoai(bmi):
 if bmi<18.5:
    return "Gầy"
 elif bmi<=24.9:
    return "Bình thường"
 elif bmi<=29.9:
    return "Hơi Béo"
 elif bmi<=34.9:
    return "Béo Phì Cấp Độ 1"
 elif bmi<=39.9:
    return "Béo Phì Cấp Độ 2"
 else:
    return "Béo Phì Cấp độ 3"
def NguyCoBenh(bmi):
 if bmi<18.5:
    return "Thấp"
 elif bmi<=24.9:
    return "Trung Bình"
 elif bmi<=29.9:
    return "Cao"
 elif bmi<=34.9:
    return "Cao"
 elif bmi<=39.9:
    return "Rất cao"
 else:
    return "Nguy Hiểm"
print("Nhập vào chiều cao:")
height=float(input())
print("Nhập vào cân nặng:")
weight=float(input())
bmi=BMI(height,weight)
print("BMI của bạn=",bmi)
print("Phân loại bạn=",PhanLoai(bmi))
print("Nguy cơ bệnh của Thím=",NguyCoBenh(bmi))
```

```python
#Bai4
def ROI(dt,cp):
 return (dt-cp)/cp
def GoiYDauTu(roi):
 if roi>=0.75:
    return "Nên đầu tư"
 else:
    return "Không nên đầu tư"
print("Chương trình tính ROI")
dt=int(input("Nhập Doanh Thu:"))
cp=int(input("Nhập chi phí:"))
roi=ROI(dt,cp)
print("Tỉ Lệ ROI=",roi)
print("==>",GoiYDauTu(roi))
```

```python
#Bai5
def fibonacci(n):
    if n<=2 :
        return 1
    return fibonacci(n-1)+fibonacci(n-2)
def listfibo(n):
    for i in range(1,n+1):
        print(fibonacci(i),end='\t')
print(fibonacci(9))
listfibo(9)
```

```python
#Bai11

def outer():
    x = 10
    y = 20
    z = 30
    m = 1
    n = 2
    def inner():
        global x
        x = x + 10
        m = 3
        n = 4
    inner()
    print("Giá trị sau khi gọi inner():")
    print("x =", x, "y =", y, "z =", z)
    print("m =", m, "n =", n)

# Trường hợp 1
print("\n--- Trường hợp 1 ---")
x = 100
print("x =", x)
print("x + 10 =", x + 10)
print("x + 20 =", x + 20)
outer()

# Trường hợp 2
print("\n--- Trường hợp 2 ---")
x = 200
print("x =", x)
print("x + 10 =", x + 10)
print("x + 20 =", x + 20)
outer()

# Trường hợp 3
print("\n--- Trường hợp 3 ---")
x = 300
print("x =", x)
print("x + 10 =", x + 10)
print("x + 20 =", x + 20)
outer()
```

```python
#Bai12
def oscillate(start, end):
    result = []
    
    current = start
    while current < 0:
        result.append(current)
        result.append(abs(current))
        current += 1

    result.append(0)
    result.append(0)

    result.append(1)
    
    limit = abs(start) + 1 
    
    for i in range(2, limit + 1):
        result.append(-i)
        result.append(i)
        
    return result


print("Kết quả chạy chương trình:")
for n in oscillate(-3, 5):
    print(n, end=' ')
print()
```
